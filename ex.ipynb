```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
import csv
```

```python
df = pd.read_csv("train/train.csv")
df["영업장명_메뉴명"].head()
```

```python
unique_values = df["영업장명_메뉴명"].unique()
unique_values_list = unique_values.tolist()
```

```python
with open("unique_영업장명_메뉴명.csv", "w", newline='', encoding='utf-8') as f:
    writer = csv.writer(f)
    for value in unique_values_list:
        writer.writerow([value])
```
